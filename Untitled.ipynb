```python
import numpy as np


def face_distance(face_encodings, face_to_compare):
    if len(face_encodings) == 0:
        return np.empty((0))

    return np.linalg.norm(face_encodings - face_to_compare, axis=1)
```

```python
known_face_encodings=[] 
lstimg=[]

know_face_names=[]

def loadImages(path):
    imagesList = listdir(path)
    for img in imagesList:
        print(path + img)
        image =  face_recognition.load_image_file(path + img)
        lstimg.append(image)
        image_encoding = face_recognition.face_encodings(image)[0]
        known_face_encodings.append(image_encoding)
        know_face_names.append(img.split('.')[0])

path = "./images/"
```

```python
loadImages(path)
```

```python
plt.figure(figsize=(10,15))
count =1
for i in range(len(lstimg)):
    plt.subplot(4,2,i+1)
    plt.imshow(lstimg[i])
    plt.axis('off')
    count=count+1
```

```python
test_image =  face_recognition.load_image_file('temp4.JPG')
face_locations = face_recognition.face_locations(test_image)
face_encodings = face_recognition.face_encodings(test_image,face_locations)
```

```python
pil_image = Image.fromarray(test_image)
```

```python
draw = ImageDraw.Draw(pil_image)
```

```python
warnings.filterwarnings("ignore") 
for(top,right,bottom,left), face_encoding in zip(face_locations,face_encodings):
    matches = face_recognition.compare_faces(known_face_encodings,face_encoding)
    name="Unknown Person"
    print(matches)
    if True in matches:
        first_match_index= matches.index(True)
        name = know_face_names[first_match_index]
    draw.rectangle(((left,top),(right,bottom)),outline=(0,0,0))
    text_width,text_height = draw.textsize(name)
    draw.rectangle(((left,bottom - text_height-10),(right,bottom)),fill=(0,0,0),outline=(0,0,0))
    draw.text((left+6,bottom - text_height -5),name,fill=(255,255,255,255))
del draw

plt.imshow(pil_image)
plt.show()
pil_image.show()
        
```

```python
cursor.execute(f'''
                SELECT * FROM CHECK_IN
                    ''')
name=[]
time=[]
day=[]
for row in cursor.fetchall():
    print(row)
    name.append(row.CNAME)
    time.append(row.CTIMEIN)
    day.append(row.CDAY)

conn.commit()
```

```python
cursor.execute(f'''
                SELECT * FROM CHECK_OUT
                    ''')
timeout=[]
for row in cursor.fetchall():
    print(row)
    timeout.append(row.CTIMEOUT)

conn.commit()
```
